### Data Visualization

Provide answer to some questions by generating charts and making use of different plotting styles.

```python
# Use countplot to determine the branch with the highest sales record
sns.countplot(x = 'Branch', data = Cities_Combined).set_title('Branch Ranking', size = 24)
sns.set(rc={'figure.figsize':(8,6)})
```

Branch A has the highest sales record

```python
# Use countplot to determine the most used payment method
sns.countplot(x = 'Payment', data = Cities_Combined).set_title('Payment Ranking', size = 24)
sns.set(rc={'figure.figsize':(8,6)})
```

The most used payment method is Epay Payment

```python
# Use countplot to determine the city with the most sales
sns.countplot(x = 'City', data = Cities_Combined).set_title('City Ranking', size = 24)
sns.set(rc={'figure.figsize':(9,7)})
```

The city with the most sales is Lagos

```python
# Explore the Payment and City Column using countplot
sns.countplot(x = 'Payment', hue = 'City', data = Cities_Combined).set_title('Payment vs City', size = 26)
sns.set(rc={'figure.figsize':(8,6)})
```

Card payment is used the most in Abuja and the least in Port Harcourt, Epay payment is used the most in Lagos and the least in Port Harcourt, and Cash payment is used the most in Port Harcout and the least in Abuja and Lagos.

```python
# Determine the highest & lowest sold product line using Countplot
sns.countplot(y = 'Product line', data = Cities_Combined).set_title('Product Line Ranking', size = 26)
sns.set(rc={'figure.figsize':(8,6)})
```

The highest sold product line is Fashion accessories while the lowest sold product line is Health and beauty.

```python
# Determine the Payment channel used by most customer to pay for each product line.
sns.countplot(y = 'Product line', hue = 'Payment', data = Cities_Combined).set_title('Payment channel for each product line. ', size = 26)
sns.set(rc={'figure.figsize':(8,6)})
```

The card payment is mostly used by customers to pay for Food and beverages. 
The Epay payment is mostly used by customers to pay for Fashion accessories. 
The Cash payment is mostly used by customers to pay for Electronic accessories. 
The Cash payment is mostly used by customers to pay for Sports and travel. 
The Epay payment is mostly used by customers to pay for Home and lifestyle. 
The Epay payment is mostly used by customers to pay for Health and beauty.

```python
# Determine the most used Payment channel for each branch.
sns.countplot(x = 'Payment', hue = 'Branch', data = Cities_Combined).set_title('Payment Channel for each branch', size = 26)
sns.set(rc={'figure.figsize':(8,6)})
```

Card payment is mostly used in Branch B, Epay payment is mostly used in Branch A and Cash payment is mostly used in Branch C.

```python
# Determine the branch with the lowest rating. 
sns.boxplot(x = 'Branch', y = 'Rating', data = Cities_Combined)
sns.set(rc={'figure.figsize':(8,6)})
```

Branch B has the lowest rating

```python
# Use catplot() to discover the way gender type affects the kind and quantity of products purchased at the supermarket
sns.catplot(x = 'Product line', y = 'Quantity', hue = 'Gender', data = Cities_Combined, aspect = 4, kind = 'box' )
sns.set(rc={'figure.figsize':(8,6)})
```

Females mostly purchase Fashion accessories and Home and lifestyle products while males mostly purchase Halth and beauty products.
